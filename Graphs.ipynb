```python
import json
import os

import matplotlib.pyplot as plt
```

```python
benches = {}
for platform in os.listdir("results"):
    with open(os.path.join("results", platform, "results.json")) as f:
        benches[platform] = json.load(f)
```

```python
for platform in sorted(benches):
    fig, ax = plt.subplots(1, len(benches[platform]), figsize=(16, 4))
    fig.suptitle(platform, fontsize="x-large")
    for i, bench_name in enumerate(benches[platform]):
        ax[i].set_title(bench_name)
        ax[i].hist(benches[platform][bench_name])
        ax[i].set_xlabel("time (ms)")
        ax[i].set_ylabel(f"events (out of {len(benches[platform][bench_name])})")
```
